```
from pyspark.sql import Window
from pyspark.sql import functions as fn
```

```
data = [
  {
    'group': 'A',
    'values' : 1
  },
  {
    'group': 'A',
    'values' : 2
  },
  {
    'group': 'B',
    'values' : 3
  },
  {
    'group': 'B',
    'values' : 3
  },
]

df = spark.createDataFrame(data)
```

# GroupBy and count

```
df_group_count = df.groupBy(
  'group'
).agg(
  fn.count('values')
)

display(df_group_count)
```

# GroupBy and count distinct

```
df_group_count_distinct = df.groupBy(
  'group'
).agg(
  fn.count_distinct('values')
)

display(df_group_count_distinct)
```

# Window and count

```
window = Window.partitionBy('group')

df_window_count = df.withColumn(
  'count', fn.count('values').over(window)
)

display(df_window_count)
```

# Workaround to Window and count distinct

```
window = Window.partitionBy('group')

df_window_count = df.withColumn(
  'count', fn.size(fn.collect_set('values').over(window))
)

display(df_window_count)
```
